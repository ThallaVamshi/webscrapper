```python
import requests
from bs4 import BeautifulSoup
```

```python
url='https://editorial.rottentomatoes.com/guide/best-horror-movies-of-all-time/'
```

```python
response=requests.get(url)
response.status_code
```

```python
soup=BeautifulSoup(response.content,'lxml')#htmp.parser
```

```python






divs=soup.find_all('div',{'class':'col-sm-18 col-full-xs countdown-item-content'})
divs
```

```python

divs[0].find('h2')
```

```python
headings=[div.find('h2') for div in divs]
headings
```

```python
[heading.text for heading in headings]
```

```python
headings[0]
```

```python
#Title
[heading.find('a')for heading in headings]
```

```python
movie_names=[heading.find('a').string for heading in headings]
movie_names
```

```python
#year
[heading.find('span',class_='subtle start-year')for heading in headings]
```

```python
years=[heading.find('span',class_='subtle start-year').string[1:-1] for heading in headings]
years
```

```python
years=[int(year)for year in years]
years
```

```python
#ratings
headings[0]
ratings=[heading.find('span',class_='tMeterScore').string for heading in headings]
```

```python
ratings=[rate.strip('%') for rate in ratings]
ratings
```

```python
ratings=[int(rate) for rate in ratings]
ratings
```

```python
#director
directors=[div.find('div',class_='info director') for div in divs]
directors
```

```python
directors[0]
```

```python
[director.find('a').string for director in directors]
```

```python
director=[director.find('a').string for director in directors]
director
```

```python
#Cast
casting=[div.find('div',class_='info cast') for div in divs]
casting
```

```python
casting[0]
```

```python
cast_links=casting[0].find_all('a')
cast_links
```

```python
cast_names=[cast.string for cast in cast_links]
cast_names
```

```python
cast=','.join(cast_names)
cast
```

```python
cast=[]
for c in casting:
  cast_links=c.find_all('a')
cast=[]
for c in casting:
  cast_links=c.find_all('a')
  cast_names=[links.string for links in cast_links]
  cast.append(','.join(cast_names))
cast
```

```python
import pandas as pd
```

```python
movies_info=pd.DataFrame()
movies_info
```

```python
movies_info['Title']=movie_names
movies_info['Year']=years
movies_info['Ratings']=ratings
movies_info['Directors']=directors
movies_info['Cast']=cast
movies_info
```

```python
from pickle import TRUE
movies_info.to_csv('movies_info.csv',index=False,header=TRUE)
```
